```python
#@title Extract the password
import requests
import re

# 1. Paste your video URL here
VIDEO_URL = "https://www.youtube.com/watch?v=hgMwe9L4tDA"

print("Fetching video data...")

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}
response = requests.get(VIDEO_URL, headers=headers)
html = response.text

# 2. Extract all text strings from the page metadata
all_simple_texts = re.findall(r'"simpleText":"(.*?)"', html)

# 3. Hunt for ALL "XX**" and "**XX" patterns and store them with their context
print("\n--- POTENTIAL PASSWORD PARTS (IN ORDER OF APPEARANCE) ---")
for text in all_simple_texts:
    # Clean up unicode characters for printing
    try:
        clean_text = text.encode().decode('unicode_escape')
    except:
        clean_text = text

    # Look for exactly two digits followed by two asterisks (e.g., "62**")
    match1 = re.search(r'(\d{2})\*\*', clean_text)
    if match1:
        part1 = match1.group(1)
        print(f"-> 1st part candidate: {part1}** (from text: '{clean_text}')")

    # Look for two asterisks followed by exactly two digits (e.g., "**87")
    match2 = re.search(r'\*\*(\d{2})', clean_text)
    if match2:
        part2 = match2.group(1)
        print(f"-> 2nd part candidate: **{part2} (from text: '{clean_text}')")
```

```python
#@title Extract all simpleText contents of the page
import requests
import re

# 1. Paste your video URL here
VIDEO_URL = "https://www.youtube.com/watch?v=hgMwe9L4tDA"

print("Fetching raw YouTube page...")
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

response = requests.get(VIDEO_URL, headers=headers)
html = response.text

# 2. Extract EVERY "simpleText" value from the entire page HTML
all_simple_texts = re.findall(r'"simpleText":"(.*?)"', html)

print(f"\nFound {len(all_simple_texts)} total text strings in the page metadata.")

# 3. Filter and print only the texts that contain numbers or the word "password"
print("\n=== 🔍 ALL TEXT STRINGS CONTAINING NUMBERS OR 'PASSWORD' ===")
for text in all_simple_texts:
    # Check if the text has a digit in it, or the word password (case-insensitive)
    if re.search(r'\d', text) or 'password' in text.lower():
        # Clean up unicode characters like \u0026 -> &
        try:
            clean_text = text.encode().decode('unicode_escape')
        except:
            clean_text = text
        print(f"-> {clean_text}")
```
