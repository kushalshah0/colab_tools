```python
#@title Extract all simpleText contents of the page
import requests
import re

# 1. Paste your video URL here
VIDEO_URL = "https://www.youtube.com/watch?v=tF1HS589pAQ"

print("Fetching raw YouTube page...")
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

response = requests.get(VIDEO_URL, headers=headers)
html = response.text

# 2. Extract EVERY "simpleText" value from the entire page HTML
all_simple_texts = re.findall(r'"simpleText":"(.*?)"', html)

print(f"\nFound {len(all_simple_texts)} total text strings in the page metadata.")

# 3. Filter and print only the texts that contain numbers or the word "password"
print("\n=== 🔍 ALL TEXT STRINGS CONTAINING NUMBERS OR 'PASSWORD' ===")
for text in all_simple_texts:
    # Check if the text has a digit in it, or the word password (case-insensitive)
    if re.search(r'\d', text) or 'password' in text.lower():
        # Clean up unicode characters like \u0026 -> &
        try:
            clean_text = text.encode().decode('unicode_escape')
        except:
            clean_text = text
        print(f"-> {clean_text}")

# 4. ULTIMATE FALLBACK: Brute force search for ANY 4-digit number in the whole page
print("\n=== 🔑 BRUTE FORCE: ALL 4-DIGIT NUMBERS FOUND IN PAGE ===")
all_4_digits = set(re.findall(r'\b\d{4}\b', html))
# Filter out common false positives like years
filtered_nums = [n for n in all_4_digits if n not in ['0000', '2020', '2021', '2022', '2023', '2024', '2025', '2026']]

for num in filtered_nums:
    print(f"🎯 {num}")
```

```python
#@title Extract the combined password
import requests
import re

# 1. Paste your video URL here
VIDEO_URL = "https://www.youtube.com/watch?v=tF1HS589pAQ"

print("Fetching video data...")
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

response = requests.get(VIDEO_URL, headers=headers)
html = response.text

# 2. Search specifically for the "1ST PASS: XX**" and "2ND PASS: **XX" patterns
match1 = re.search(r'1ST\s*PASS[:\s]*(\d{2})\*\*', html, re.IGNORECASE)
match2 = re.search(r'2ND\s*PASS[:\s]*\*\*(\d{2})', html, re.IGNORECASE)

if match1 and match2:
    # Extract the two parts
    part1 = match1.group(1)
    part2 = match2.group(1)

    # Combine them
    final_password = part1 + part2

    print("\n=== 🔑 PASSWORD EXTRACTED SUCCESSFULLY ===")
    print(f"First part found:  {part1}**")
    print(f"Second part found: **{part2}")
    print("-" * 40)
    print(f"👉 FINAL PASSWORD: {final_password}")
    print("-" * 40)
else:
    # Fallback just in case the creator typed it slightly differently
    print("\nCould not find the exact '1ST/2ND PASS' pattern. Printing all text with 'PASS' in it:")
    all_texts = re.findall(r'"simpleText":"(.*?)"', html)
    for t in all_texts:
        if 'pass' in t.lower():
            try:
                print("->", t.encode().decode('unicode_escape'))
            except:
                print("->", t)
```
